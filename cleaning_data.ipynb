```python
hitters_file = pd.read_csv('hitter_stats_metrics.csv')
hitters_file[['last_name', 'first_name']] = hitters_file['last_name, first_name'].str.split(',', n=1, expand=True)
hitters_file.drop(columns=['last_name, first_name'], inplace=True)
hitters_file.columns = hitters_file.columns.str.replace('_',' ')
hitters_file.head()
```

```python
hitters_file.to_csv('cleaned_hitter_stats_metrics.csv', index=False)
```

```python
pitchers_file = pd.read_csv('pitcher_stats_metrics.csv')
pitchers_file[['last_name', 'first_name']] = pitchers_file['last_name, first_name'].str.split(',', n=1, expand=True)
pitchers_file.drop(columns=['last_name, first_name'], inplace=True)
pitchers_file.head()
```
